```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.io.json import json_normalize
```

```python
import pathlib
path = str(pathlib.Path().absolute())
```

```python
df=pd.read_json(path+'/pasos.json')
df.head()
```

Cantidad de pasos fronterizos por provincias. No hay información de contexto para los datos. Debemos suponer a partir de la examinación, que se incluyen aeropuertos como "pasos fronterizos". De otra manera, ¿qué paso fronterizo tendría Santiago del Estero?

```python
df.province.value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.set_context('talk')
ax=sns.countplot(df.province)
ax.set(xlabel='Provincia',ylabel='Pasos fronterizos',title='Pasos fronterizos por provincia')
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show(ax)
```
